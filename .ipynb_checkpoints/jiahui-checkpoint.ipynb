### 4.3 Plot `Verbal`, `Math`, and `Rate` appropriately on the same boxplot chart

Think about how you might change the variables so that they would make sense on the same chart. Explain your rationale for the choices on the chart. You should strive to make the chart as intuitive as possible. 


````python
import numpy as np

def zval(series): #normalizes a series
    try: 
        sf = pd.Series([(x-np.mean(series)) / np.std(series) for x in series])
        return sf
    except:
        series2 = []
        for i in range(len(series)):
            if series[i]!='-':  #to handle the '-' entries 
                series2.append(series[i])
        sf = pd.Series([(x-np.mean(series2)) / np.std(series) for x in series2])
        return sf
    
norm_verbal = zval(sat['Verbal'])
norm_math = zval(sat['Math'])
norm_rate = zval(sat['Rate'])
combined = pd.DataFrame({'norm_verbal': norm_verbal,
                        'norm_math':norm_math,
                        'norm_rate':norm_rate})
#combined = [(x-x.mean())/x.std() for x in sat.loc[:,['Verbal','Math','Rate']]]
combined.plot.box()
````

````python
# Calculate point estimates

aMed = np.median(drug_freq['alcohol-frequency'])
hMed = np.median(drug_freq['marijuana-frequency'])
avar = np.var(drug_freq['alcohol-frequency'], ddof=1)
hvar = np.var(drug_freq['marijuana-frequency'], ddof=1)
length = (np.size(drug_freq['alcohol-frequency']))

print aMed, hMed, astd, hstd, length
````

````python
from scipy.stats import t

# Calculate t statistic.
ts = (hMed-aMed) / ((avar/length)+(hvar/length))**0.5

# Calculate p value.
pv = t.sf(np.abs(ts), length) * 2 

print ts, pv
````

Because the p-value (0.565) is much larger than the confidence interval of 0.01, we reject the null hypothesis.

<p style="color:#ff0000;">
<b>Important note!</b> Null hypothesis is only rejected when p-value is <b>smaller</b> than the <b>significance level</b> (aka alpha aka 0.01). When alpha is 0.01, <b>confidence level</b> = 0.99 (99%). Confidence interval != confidence level. 2. Describe clearly in words the process of calculating the spearman rank correlation.
  - Hint: the word "rank" is in the name of the process for a reason!


````python
stats.spearmanr(sat['Verbal'], sat['Math']) 
#extremely high degree of freedom. is statistically significant.
````

Explanation of spearman rank correlation calculation

1. Under each variable (or column), each statistic is ranked according to its position compared to other statistics. 
2. Each rank is subtracted from the previous statistic's rank, to give a difference, d. 
3. d is squared and summed, and then put into the Spearman Rank formula.
